 2. 描述算法功能，并判断是否最高效

该算法的功能是对数组进行排序。它实际上是经典的 **Stooge Sort，愚蠢排序**。

算法思想是：

1. 排序前 $2/3$ 的元素；
2. 排序后 $2/3$ 的元素；
3. 再排序前 $2/3$ 的元素。

它可以完成排序，但不是最高效的排序算法。

原因是它的时间复杂度为：

$$
\Theta(n^{2.71})
$$

而归并排序、堆排序、快速排序平均情况下都可以达到：

$$
O(n\log n)
$$

所以该算法不是最高效的解决方案。

##
 3. Python 编程实现


````python

def stooge_sort(arr, l=0, r=None):
    if r is None:
        r = len(arr) - 1

    if l >= r:
        return arr

    if arr[l] > arr[r]:
        arr[l], arr[r] = arr[r], arr[l]

    n = r - l + 1

    if n >= 3:
        m = (2 * n + 2) // 3   # ceil(2n/3)

        stooge_sort(arr, l, l + m - 1)
        stooge_sort(arr, r - m + 1, r)
        stooge_sort(arr, l, l + m - 1)

    return arr


A = [5, 2, 9, 1, 7, 3]
print("排序前:", A)
print("排序后:", stooge_sort(A))

# 时间复杂度: O(n^2.71)
````

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:
1. 设计算法计算最大利润和买入、卖出日期

### 
算法思路:

最直接的方法是枚举所有买入日期和卖出日期：

1. 固定买入日期 $i$；
2. 枚举卖出日期 $j$；
3. 计算区间 $profits[i..j]$ 的总收益；
4. 记录最大收益和对应日期。

### 
时间复杂度:

如果每次区间求和都重新计算，复杂度是 $O(n^3)$。

可以在枚举右端点时累加当前区间和，这样复杂度降为：

$$
O(n^2)
$$

空间复杂度：

$$
O(1)
$$

---

## 
2. 设计一个 $O(n)$ 算法

### 
算法思路:

使用 Kadane 算法。

核心思想：

- 如果当前累计收益小于 0，就没有必要继续持有之前的区间；
- 重新从当前天开始计算；
- 一边遍历，一边维护最大收益。

时间复杂度：

$$
O(n)
$$

空间复杂度：

$$
O(1)
$$


````python

def max_profit_bruteforce(profits):
    n = len(profits)
    max_sum = float("-inf")
    buy_day = sell_day = 0

    for i in range(n):
        current_sum = 0
        for j in range(i, n):
            current_sum += profits[j]
            if current_sum > max_sum:
                max_sum = current_sum
                buy_day = i + 1
                sell_day = j + 1

    return max_sum, buy_day, sell_day


def max_profit_kadane(profits):
    max_sum = float("-inf")
    current_sum = 0

    start = 0
    buy_day = sell_day = 0

    for i, x in enumerate(profits):
        if current_sum <= 0:
            current_sum = x
            start = i
        else:
            current_sum += x

        if current_sum > max_sum:
            max_sum = current_sum
            buy_day = start + 1
            sell_day = i + 1

    return max_sum, buy_day, sell_day


profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]

print("O(n^2)算法:", max_profit_bruteforce(profits))
print("O(n)算法:", max_profit_kadane(profits))

# O(n^2)算法时间复杂度: O(n^2)
# O(n)算法时间复杂度: O(n)
````

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:
1. 设计 $O(\log n)$ 算法找到一个局部最小值

### 
算法思路:

因为数组两端满足：

- 左边趋势一开始是下降的；
- 右边趋势最后是上升的；

所以数组中一定存在至少一个局部最小值。

可以用二分查找：

设当前检查位置为 `mid`：

1. 如果 `A[mid]` 同时小于左右邻居，则它就是局部最小；
2. 如果 `A[mid] > A[mid-1]`，说明左边存在下降趋势，局部最小在左侧；
3. 否则局部最小在右侧。

每次排除一半区间，因此时间复杂度为：

$$
O(\log n)
$$

空间复杂度：

$$
O(1)
$$

---

## 
2. 找出所有局部最小值

###
 算法思路:

要找出所有局部最小值，必须检查每个内部元素是否小于左右邻居。

因此需要遍历一次数组。

时间复杂度：

$$
O(n)
$$

空间复杂度：

$$
O(k)
$$

其中 $k$ 是局部最小值的个数。


````python

def find_one_local_min(A):
    n = len(A)
    left, right = 1, n - 2

    while left <= right:
        mid = (left + right) // 2

        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]

        elif A[mid] > A[mid - 1]:
            right = mid - 1

        else:
            left = mid + 1

    return None


def find_all_local_mins(A):
    result = []

    for i in range(1, len(A) - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            result.append(A[i])

    return result


A = [9, 3, 7, 2, 1, 4, 5]

print("一个局部最小值:", find_one_local_min(A))
print("所有局部最小值:", find_all_local_mins(A))

# find_one_local_min 时间复杂度: O(log n)
# find_all_local_mins 时间复杂度: O(n)
````

## 
3. 数据量超过计算机存储能力时，找前 $k$ 小

###
   算法思路:

因为计算机一定能存储 $k$ 个数据，所以可以维护一个大小为 $k$ 的大根堆。

步骤：

1. 先读入前 $k$ 个数，建立大根堆；
2. 后续每读入一个数 $x$：
   - 如果 $x$ 大于等于堆顶，说明它不可能属于前 $k$ 小，丢弃；
   - 如果 $x$ 小于堆顶，则弹出堆顶，把 $x$ 插入堆中；
3. 所有数据读完后，堆中保存的就是前 $k$ 小；
4. 最后排序输出即可。

时间复杂度：

$$
O(n\log k+k\log k)
$$

通常写作：

$$
O(n\log k)
$$

空间复杂度：

$$
O(k)
$$


````python

import heapq


def k_smallest_by_sort(arr, k):
    return sorted(arr)[:k]


def k_smallest_by_min_heap(arr, k):
    heap = []

    for x in arr:
        heapq.heappush(heap, x)

    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap))

    return result


def k_smallest_stream(stream, k):
    max_heap = []

    for x in stream:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -x)
        else:
            if x < -max_heap[0]:
                heapq.heapreplace(max_heap, -x)

    result = [-x for x in max_heap]
    result.sort()
    return result


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3

print("排序法:", k_smallest_by_sort(arr, k))
print("小根堆法:", k_smallest_by_min_heap(arr, k))
print("大数据流法:", k_smallest_stream(arr, k))

# 排序法时间复杂度: O(n log n)
# 小根堆法时间复杂度: O((n+k) log n)
# 大数据流法时间复杂度: O(n log k)
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:
##
星球碰撞问题。
##
数组中：
- 正数表示星球自左向右运动；
- 负数表示星球自右向左运动；
- 绝对值表示星球质量；
- 只有左边是正数、右边是负数时才会相撞；
- 质量小的消失；
- 质量相同时，负数星球消失，正数星球保留。

---

## 
算法设计思路:

使用栈模拟星球运动。

依次扫描数组中的星球：

1. 如果当前星球不会和栈顶星球碰撞，直接入栈；
2. 如果栈顶是正数，当前星球是负数，则发生碰撞；
3. 比较两者质量：
   - 栈顶质量小：栈顶消失，继续和新的栈顶比较；
   - 栈顶质量大：当前星球消失；
   - 质量相同：根据题目规则，负数星球消失，正数星球保留。

每个星球最多入栈一次、出栈一次。

因此时间复杂度：

$$
O(n)
$$

空间复杂度：

$$
O(n)
$$


````python

def asteroid_collision(A):
    stack = []

    for x in A:
        alive = True

        while alive and stack and stack[-1] > 0 and x < 0:
            if abs(stack[-1]) < abs(x):
                stack.pop()
            elif abs(stack[-1]) > abs(x):
                alive = False
            else:
                # 质量相同：负数星球消失，正数星球保留
                alive = False

        if alive:
            stack.append(x)

    return stack


A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
A2 = [23, -8, 9, -3, -7, 9, -23, 22]

print(asteroid_collision(A1))
print(asteroid_collision(A2))

# 时间复杂度: O(n)
# 空间复杂度: O(n)
````

$$
---

## 
算法设计思路:

这是 0-1 背包问题。每块大理石只能选择拿或者不拿，不能拆分。

使用动态规划。

定义：

$$
dp[i][w]
$$

表示只考虑前 $i$ 个物品，在容量为 $w$ 的情况下可以获得的最大价值。

状态转移：

如果第 $i$ 个物品重量大于当前容量：

$$
dp[i][w]=dp[i-1][w]
$$

否则：

$$
dp[i][w]=\max(dp[i-1][w],\ dp[i-1][w-weight_i]+value_i)
$$

---

## 
复杂度分析:

设物品数量为 $n$，背包容量为 $W$。

时间复杂度：

$$
O(nW)
$$

空间复杂度：

$$
O(nW)
$$

如果使用一维滚动数组，可以优化到：

$$
O(W)
$$

---

## 
本题答案:

选择重量为 4、价值为 40 的大理石，以及重量为 3、价值为 50 的大理石。

总重量：

$$
4+3=7\le 9
$$

总价值：

$$
40+50=90
$$

所以最大价值为：

$$
\boxed{90}
$$


````python

def knapsack_01(items, capacity):
    n = len(items)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight, value = items[i - 1]

        for w in range(capacity + 1):
            if weight > w:
                dp[i][w] = dp[i - 1][w]
            else:
                dp[i][w] = max(
                    dp[i - 1][w],
                    dp[i - 1][w - weight] + value
                )

    return dp[n][capacity]


items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9

print("最大价值:", knapsack_01(items, capacity))

# 时间复杂度: O(nW)
# 空间复杂度: O(nW)
````

## 问题 8

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

<div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:
## 
算法设计思路:

二叉搜索树的一个重要性质：

> 对二叉搜索树进行中序遍历，得到的序列应该是严格递增的。

如果有两个节点被交换，那么中序遍历序列中会出现逆序。

例如：

```text
1, 3, 2, 4
```

逆序位置是：

```text
3 > 2
```

需要交换 3 和 2。

如果交换的两个节点不相邻，例如：

```text
1, 4, 3, 2, 5
```

会出现两次逆序：

```text
4 > 3
3 > 2
```

此时应该交换第一次逆序中的较大值 4，和第二次逆序中的较小值 2。

### 
算法步骤:

1. 中序遍历二叉搜索树；
2. 用 `prev` 记录上一个访问节点；
3. 如果发现 `prev.val > current.val`：
   - 第一次发现时，记录 `first = prev`；
   - 每次发现时，记录 `second = current`；
4. 遍历结束后，交换 `first.val` 和 `second.val`。

---

## 
复杂度分析:

每个节点访问一次。

时间复杂度：

$$
O(n)
$$

递归栈空间复杂度：

$$
O(h)
$$

其中 $h$ 是树的高度。


````python

from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def recover_tree(root):
    first = second = prev = None

    def inorder(node):
        nonlocal first, second, prev

        if not node:
            return

        inorder(node.left)

        if prev and prev.val > node.val:
            if first is None:
                first = prev
            second = node

        prev = node

        inorder(node.right)

    inorder(root)

    if first and second:
        first.val, second.val = second.val, first.val

    return root


def build_tree(values):
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1

    return root


def tree_to_list(root):
    if not root:
        return []

    result = []
    queue = deque([root])

    while queue:
        node = queue.popleft()

        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)

    while result and result[-1] is None:
        result.pop()

    return result


root = build_tree([1, 3, None, None, 2])
recover_tree(root)

print(tree_to_list(root))

# 时间复杂度: O(n)
# 空间复杂度: O(h)
````

## 问题 9 \*
给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。



<div style="text-align:center;width:800px;"> <img alt="" src="./fig/hw3q6.png"> </div>

请分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:
## 
算法设计思路:

对每个位置 $i$ 来说，它能接的水量取决于：

1. 左边最高柱子高度；
2. 右边最高柱子高度；
3. 当前柱子高度。

位置 $i$ 能接的水为：

$$
\min(leftMax_i,\ rightMax_i)-height_i
$$

如果结果小于 0，则该位置不能接水。

可以使用双指针优化空间。

### 
双指针思路:

1. 设置左右指针 `left` 和 `right`；
2. 维护 `left_max` 和 `right_max`；
3. 哪一侧较低，就先处理哪一侧；
4. 因为较低一侧的接水上限已经确定。

---

## 
复杂度分析:

每个元素最多被访问一次。

时间复杂度：

$$
O(n)
$$

空间复杂度：

$$
O(1)
$$


````python

def trap_rain_water(height):
    if not height:
        return 0

    left, right = 0, len(height) - 1
    left_max = right_max = 0
    water = 0

    while left < right:
        if height[left] < height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                water += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                water += right_max - height[right]
            right -= 1

    return water


height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]

print("接水量:", trap_rain_water(height))

# 时间复杂度: O(n)
# 空间复杂度: O(1)
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:
## 
算法设计思路:

使用深度优先搜索 DFS。

从根节点向下遍历时，维护当前路径已经形成的数字 `current`。

假设当前节点值为 `node.val`，则更新：

$$
current = current \times 10 + node.val
$$

如果当前节点是叶子节点，则将 `current` 加入答案。

---

## 
例子:

路径：

```text
1 -> 2 -> 3
```

表示数字：

$$
123
$$

计算方式：

$$
((1 \times 10 + 2)\times 10)+3=123
$$

---

## 
复杂度分析:

每个节点访问一次。

时间复杂度：

$$
O(n)
$$

递归栈空间复杂度：

$$
O(h)
$$

其中 $h$ 为树的高度。


````python

from collections import deque


class TreeNode2:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def sum_numbers(root):
    def dfs(node, current):
        if not node:
            return 0

        current = current * 10 + node.val

        if not node.left and not node.right:
            return current

        return dfs(node.left, current) + dfs(node.right, current)

    return dfs(root, 0)


def build_tree2(values):
    if not values or values[0] is None:
        return None

    root = TreeNode2(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode2(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode2(values[i])
            queue.append(node.right)
        i += 1

    return root


root1 = build_tree2([1, 2, 3])
root2 = build_tree2([4, 9, 0, 5, 1])

print("示例1:", sum_numbers(root1))
print("示例2:", sum_numbers(root2))

# 时间复杂度: O(n)
# 空间复杂度: O(h)
````
